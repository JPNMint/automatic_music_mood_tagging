```python
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
import os
src = '/home/ykinoshita/humrec_mood_tagger/output_fold/FCN_6_layer'
file_names = os.listdir(src)
results_r2 = pd.DataFrame()
results_rmse = pd.DataFrame()

for file_name in file_names:
    if "No_oversampling_" in file_name:
        header = f"DAug_{file_name[-1]}"
    elif "average_density_oversampling" in file_name:
        if "average_density_oversampling_V2" in file_name:
            header = f"DOMR_percentile_{file_name[-3:]}" 
        else:
            
            header = f"DOMR_{file_name[-3:]}"
    elif "density_oversampling_augmentation" in file_name:
        if "density_oversampling_augmentation_V2" in file_name:
            header = f"DOMR+_percentile{file_name[-5:]}" 
        else:
            header = f"DOMR+_{file_name[-5:]}" 
    else:
        header = file_name
    csv_path = (src+'/'+file_name+'/fold_metrics_detailed.csv')
    results_r2[f'{header}'] = pd.read_csv(csv_path)['R2']
    results_rmse[f'{header}'] = pd.read_csv(csv_path)['RMSE']
    #shutil.copy2(os.path.join(src_path, file_name), dst_path)
results_r2 = results_r2.reindex(sorted(results_rmse.columns), axis=1)
results_rmse = results_rmse.reindex(sorted(results_rmse.columns), axis=1)
```

```python
results_r2.to_csv(f'/home/ykinoshita/humrec_mood_tagger/output_fold/metrics_r2.csv', index=False)
```

```python
results_rmse.to_csv(f'/home/ykinoshita/humrec_mood_tagger/output_fold/metrics_rmse.csv', index=False)
```
